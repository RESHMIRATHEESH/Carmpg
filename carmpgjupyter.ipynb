```python
df.drop("car_name",axis=1,inplace=True)
```

```python
df
```

```python
df["origin"]=df.origin.replace({1:"america",2:"europe",3:"asia"})
```

```python
df.head()
```

```python
df=pd.get_dummies(df,columns=["origin"])
```

```python
df.head()
```

```python
bool_columns=['origin_america','origin_asia','origin_europe']
df[bool_columns]=df[bool_columns].astype(int)
```

```python
df.head()
```

```python
df.describe().transpose()
```

```python
temp=pd.DataFrame(df.hp.str.isdigit())
```

```python
temp[temp["hp"]==False]
```

```python
df["hp"]=df.hp.replace({"?":np.nan})
```

```python
df[df.isnull().any(axis=1)]
```

```python
df=df.apply(lambda x:x.fillna(x.median(),axis=0))
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df["hp"]=df["hp"].astype(int)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X=df.drop("mpg",axis=1)
```

```python
Y=df[["mpg"]]
```

```python
X=X.drop(["origin_america","origin_europe","origin_asia"],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
for i ,j in enumerate(x_train.columns):
    print("The co-efficient of {} is {}".format(j,reg.coef_[0][i]))
```

```python
reg.coef_
```

```python
reg.intercept_[0]
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
dff=pd.concat([x_train,y_train],axis=1)
```
